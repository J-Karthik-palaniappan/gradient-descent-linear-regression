```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.random((1000))*100
y = np.random.random((1000))*100
```

## Gradient descent

```python
m = 0
c = 0
error = []
for i in range(50000): 
    y1 = m*x + c
    
    error.append(((y-y1)**2).mean())
    
    m -= 0.0001 * (-2/len(x)) * sum(x * (y - y1))
    c -= 0.0001 * (-2/len(x)) * sum(y - y1)
```

```python
plt.scatter(x,y)
plt.plot(x,m*x+c,color='orange')
```

```python
plt.plot(range(len(error)),error)
```

```python
print("rmse:",np.sqrt(((y - (m*x+c))**2).mean()))
```

## SKLearn method

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
x = np.reshape(x,(1000,1))
reg.fit(x, y)
ypred = reg.predict(x)
plt.scatter(x,y)
plt.plot(x,ypred,color='orange')
```

```python
print("rmse:",np.sqrt(((y - ypred)**2).mean()))
```

## Comparison

```python
print("sklerarn:",reg.coef_[0],reg.intercept_)
print("gradient descent:",m,c)
```

#### Both of them have almost same values of intercept and slope but gradient descent takes much longer time than sklearn because it does not use the direct formula but keeps optimizing the parameters in every iteration
